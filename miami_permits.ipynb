# Miami Permits

```python
# import pandas
import pandas as pd

# import DecisionTree ML model
from sklearn.tree import DecisionTreeRegressor

# pull CSV file from google drive
miami_data_filepath = '/content/drive/MyDrive/csv/miami_permits.csv'

# read CSV file
miami_data = pd.read_csv(miami_data_filepath)

# print a summary of the data in miami_data
miami_data.describe()

```

```python
miami_data.columns
```

```python
# dropping missing data 
# miami_data = miami_data.data.dropna(axis=0)
miami_data.dropna(inplace=True)
```

```python
# selecting the prediction target
y = miami_data.DaysInCityNumeric

# setting features which are columns that determine the daysincity
miami_features = ['NewAdditionCost', 'TotalCost', 'TotalSQFT']
X = miami_data[miami_features]

X.describe()
X.head

# define model. Specify a random_state number to ensure the same results each run
miami_model = DecisionTreeRegressor(random_state=1)

# fit model
miami_model.fit(X,y)

print("Making predictions on the following 5 permits")
print(X.head())
print("The predictions are")
print(miami_model.predict(X.head()))
print(miami_model.predict(X))
```
